```python
#count of loans by it's purpose
LendingClubData[['purpose_car','purpose_credit_card','purpose_debt_consolidation','purpose_educational',
                'purpose_home_improvement','purpose_house','purpose_major_purchase','purpose_medical',
                'purpose_moving','purpose_other','purpose_renewable_energy','purpose_small_business',
                'purpose_vacation','purpose_wedding']].sum()
```

```python
#list of top 10 loans loan amounts & their purposes
LendingClubData[['loan_amnt','purpose_car','purpose_credit_card','purpose_debt_consolidation',
                 'purpose_educational','purpose_home_improvement','purpose_house','purpose_major_purchase',
                 'purpose_medical','purpose_moving','purpose_other','purpose_renewable_energy',
                 'purpose_small_business','purpose_vacation','purpose_wedding']
               ].nlargest(10,'loan_amnt')
```

```python
#attempting to melt the table into something better with the power of google
PurposeRows = [col for col in LendingClubData.columns if 'purpose' in col]
melted_df=LendingClubData.melt(id_vars=['loan_amnt'], value_vars=PurposeRows)
melted_df.drop(melted_df[melted_df['value']==0.0].index,inplace=True)
melted_df.nlargest(10,'loan_amnt')
```
